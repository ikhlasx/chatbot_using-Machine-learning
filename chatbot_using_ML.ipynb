```python
import nltk
import random
import numpy as np
import pandas as pd
import string
import warnings
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.linear_model import LogisticRegression

nltk.download('punkt')
nltk.download('wordnet')
```

```python
data= {
    "text" :["hello","hai","hi","good morning","good night","hey","how are you","see you","what is your name","what do you do","what is the technique used to build you","who are you","thank you","see you bye","iam going","bye bye","so good","see you later","nice to see you"],
    "intent":["greeting","greeting","greeting","greeting","greeting","greeting","greeting","greeting","about","about","about","about","farewell","farewell","farewell","farewell","gratitude","gratitude","gratitude"]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
def preprocess(data):
  tokens=word_tokenize(data.lower())
  tokens = [word for word in tokens if word.isalnum()]
  return " ".join(tokens)
df["clean_text"]= df["text"].apply(preprocess)
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
x_train_tfidf = vectorizer.fit_transform(df["clean_text"])
```

```python
model = LogisticRegression()
model.fit(x_train_tfidf, df["intent"])
```

```python
model.predict(x_train_tfidf)
```

```python
def chatbot_response(user_input):
  user_input = preprocess(user_input)
  user_tfidf = vectorizer.transform([user_input])

  similarity_score = cosine_similarity(user_tfidf,x_train_tfidf)
  max_similarity = np.max(similarity_score)
  #best_match_index = np.argmax(similarity_score)
  print(max_similarity)


  if max_similarity < 0.2:
    return "sorry I  didn't understand that."
#predict intent using logistic reegression
  predicted_intent = model.predict(user_tfidf)[0]


  responses = {
      "greeting": "hello, how can i help you",
      "about" : "iam a chatbot build using logistic regression and cosine similarity!",
      "farewell" : "goodbye have a great day",
      "gratitude":"You'r welcome",
  }

  return responses.get(predicted_intent,"sorry I didn't understand that.")
```

```python
while True:
  user_input = input("you : ")
  if user_input.lower() in ["bye", "quit"]:
    print("chatbot: goodbye")
    break
  print("chatbot: ",chatbot_response(user_input))
```
